```python
import pandas as pd
```

```python
df = pd.read_csv('rules.csv')
```

```python
df.head()
```

```python
def clean(row):
    row['Antecedents'] = row['Antecedents'].replace("{", "").replace("}", "").replace("'", "").split(',')

    row['Consequents'] = row['Consequents'].replace("{", "").replace("}", "").replace("'", "").split(',')

    return row
```

```python
df = df.apply(clean, axis=1)
```

```python
df.head()
```

```python
df.to_csv('association_rules.csv', index=False) 
```
